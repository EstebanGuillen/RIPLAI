```python
def create_image(path,target_directory):
    img  = tf.io.read_file(path)
    img = tfio.image.decode_dicom_image(img, scale='auto', on_error='lossy', dtype=tf.dtypes.uint8)
    img = tf.image.convert_image_dtype(img, tf.float32)
    img = tf.image.resize_with_pad(img, IMG_WIDTH, IMG_HEIGHT)
    img = tf.squeeze(img)
    img = tf.expand_dims(img,-1)
    
    pixel_values = np.squeeze(img.numpy())
    rescaled = (255.0 / pixel_values.max() * (pixel_values - pixel_values.min())).astype(np.uint8)
    im = Image.fromarray(rescaled,'L')
    
    file_name = os.path.basename(path)
    
    file_name_png = file_name.replace('.dcm','.png')
    im.save(target_directory + file_name_png)
```

```python
import os

for path in true_positives_train:
    try:
        create_image(path,positive_target_train_directory)
    except:
        print('error processing', path)
    
```

```python
for path in true_positives_validation:
    try:
        create_image(path,positive_target_validation_directory)
    except:
        print('error processing', path)
    
```
