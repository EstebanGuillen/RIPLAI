```python


#path = '/opt/AIStorage/PLAYGROUND/images/1024/'

path = '/opt/AIStorage/PLAYGROUND/images/512/filtered/data_filtered/'


data = ImageDataBunch.from_folder(path, size=512, valid_pct=0.2,bs=8, ds_tfms=(rand_pad(2, 512), []))


epochs=4

#arch = models.densenet121

#arch = models.densenet169

arch = models.densenet201

    
fine_tune_epochs = 6
```

```python
learn = cnn_learner(data, arch, metrics=accuracy, bn_final=False)
#learn = cnn_learner(data, arch, metrics=accuracy, bn_final=True).to_fp16()
#learn = cnn_learner(data, arch, metrics=accuracy, bn_final=True)
#learn = cnn_learner(data, arch, metrics=accuracy, bn_final=False,opt_func=optim.SGD).to_fp16()
#learn = cnn_learner(data, arch, metrics=accuracy, bn_final=False,opt_func=optim.SGD)
```

```python
#learn.lr_find()
#learn.recorder.plot()
```

```python
lr = 3e-3
```

```python
learn.model = nn.DataParallel(learn.model)
#learn.fit_one_cycle(epochs)
learn.fit(epochs)
```

```python
learn.recorder.plot_losses()
```

```python
learn.recorder.plot_metrics()
```

```python
learn.recorder.plot_lr(show_moms=True)
```

```python
learn.freeze_to(-2)
#learn.fit_one_cycle(fine_tune_epochs, lr/10)
learn.fit(fine_tune_epochs, lr/10)
```

```python
learn.recorder.plot_losses()
```

```python
learn.recorder.plot_metrics()
```

```python
learn.recorder.plot_lr(show_moms=True)
```
